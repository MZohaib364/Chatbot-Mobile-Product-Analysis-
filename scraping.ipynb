```python
driver = webdriver.Chrome()
```

```python
url = 'https://www.daraz.pk/'
driver.get(url)
```

```python
banner = driver.find_element(By.CLASS_NAME, 'lzd-site-nav-menu-dropdown')
e_devices = banner.find_element(By.XPATH, '/html/body/div[1]/div/div/div[2]/div/div/div/div/div/div/div/ul/li[7]')
phones = e_devices.find_element(By.XPATH, '/html/body/div[1]/div/div/div[2]/div/div/div/div/div/div/div/ul/ul[7]/li[1]')
anchor_tag = phones.find_element(By.TAG_NAME, 'a')
main_url = anchor_tag.get_attribute('href')
```

```python
driver.get(main_url)
```

```python
#Scroll and Then Run this Cell

page_links = []

p1 = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[3]/div[2]/div/ul/li[2]')
anchor_tag = p1.find_element(By.TAG_NAME, 'a')
plink = anchor_tag.get_attribute('href')
page_links.append(plink)

p2 = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[3]/div[2]/div/ul/li[3]')
anchor_tag = p2.find_element(By.TAG_NAME, 'a')
plink = anchor_tag.get_attribute('href')
page_links.append(plink)

p3 = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[3]/div[2]/div/ul/li[4]')
anchor_tag = p3.find_element(By.TAG_NAME, 'a')
plink = anchor_tag.get_attribute('href')
page_links.append(plink)

p4 = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[3]/div[2]/div/ul/li[5]')
anchor_tag = p4.find_element(By.TAG_NAME, 'a')
plink = anchor_tag.get_attribute('href')
page_links.append(plink)

p5 = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[3]/div[2]/div/ul/li[6]')
anchor_tag = p5.find_element(By.TAG_NAME, 'a')
plink = anchor_tag.get_attribute('href')
page_links.append(plink)
 
```

```python
page_links
```

```python
import time

links = []

for j in range(len(page_links)):
    driver.get(page_links[j])
    time.sleep(2)
    ad_grid = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[4]/div/div/div[1]/div[2]')
    ad_cards = ad_grid.find_elements(By.CLASS_NAME, 'gridItem--Yd0sa')
    
    for i in range(len(ad_cards)):
        anchor_tag = ad_cards[i].find_element(By.TAG_NAME, 'a')
        link = anchor_tag.get_attribute('href')
        links.append(link)
```

```python
name = []
price = []
company = []
rev_list = []
rev_id = []
specifications = []
rating = []
id = []

# driver.get(links[0])
```

```python
import time

exclude_keywords = ['case', 'cover', 'charger']

counter_id = 0

for i in range(124,len(links)):
    driver.get(links[i])
    time.sleep(2)
    
    try:
        container = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.ID, 'block-TdCYL2moZe'))
            )


    #     container = driver.find_element(By.ID, 'block-TdCYL2moZe')
        name_temp = container.find_element(By.CLASS_NAME, 'pdp-product-title')
        product_name = name_temp.find_element(By.XPATH, '/html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[3]/div/div/span').text

        if any(keyword in product_name.lower() for keyword in exclude_keywords):
                print(f'Skipping product {product_name} due to exclusion keywords.')
                continue

        name.append(product_name)
    except TimeoutException:
        name.append('-')
        pass
    
    counter_id += 1
    id.append(counter_id)
    
    
    price_temp = driver.find_element(By.CLASS_NAME, 'pdp-mod-product-price')
    price.append(price_temp.find_element(By.XPATH, '/html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[7]/div/div/span').text)

    brand_temp = driver.find_element(By.ID, 'module_product_brand_1')
    company.append(brand_temp.find_element(By.TAG_NAME, 'a').text)

    driver.execute_script("window.scrollTo(0, 500);")
    time.sleep(1)

    try:
        rating_temp = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, 'review-info-left'))
            )
        try:
    #         rating_temp = driver.find_element(By.CLASS_NAME, 'review-info-left')
            rating.append(rating_temp.find_element(By.XPATH, '/html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div/div[2]/div[1]/div[1]/div[1]/span[1]').text)
        except NoSuchElementException:
            rating.append('-')
    except TimeoutException:
        rating.append('-')
        pass

    driver.execute_script("window.scrollTo(500, 1000);")
    time.sleep(1)

    l1 = []
    l2 = []

    for page in range(0,5):
            try:
                reviews_container = driver.find_element(By.CLASS_NAME, 'review-content')
            except NoSuchElementException:
                reviews.append("-")
                print('No Reviews')
                break

            reviews = reviews_container.find_elements(By.CLASS_NAME,'review-item')


            for i in range(len(reviews)):
                l1.append(reviews[i].find_element(By.CLASS_NAME, 'review-content-sl').text)
                l2.append(reviews[i].find_element(By.CLASS_NAME, 'user').text)

            try:
                if(page != 4):
                    nxt_button = reviews_container.find_element(By.XPATH, '/html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div/div[3]/div[4]/ul/li[9]/button')
                    driver.execute_script("arguments[0].click();", nxt_button)
                    time.sleep(2)
            except NoSuchElementException:
                print('No More Pages')
                break


    rev_list.append(l1)
    rev_id.append(l2)

    driver.execute_script("window.scrollTo(1000, 1500);")
    time.sleep(1)

    specs_container = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, 'module_product_detail'))
        )
#     specs_container = driver.find_element(By.ID,'module_product_detail')
    specifications.append(specs_container.find_element(By.XPATH,'/html/body/div[4]/div/div[9]/div[1]/div[2]/div/div/div[1]/div[1]').text.split('\n'))


```

```python
import pandas as pd
dict = {
    'Name': name,
    'Price': price,
    'Company': company,
#     'Review Content': rev_list,
#     'Review ID': rev_id,
    'Specifications': specifications,
    'Rating': rating,
    'Product ID' : id
}

df1 = pd.DataFrame(dict)

df1.set_index('Product ID', inplace=True)

df1.to_csv('test_data.csv')
```
